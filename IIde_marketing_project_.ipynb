# Find batch wise bifurcation of hired students How many students were hired in the following companies - IIDE, Kinnect, GroupM & Schbang.

```python
hired_students = data[data['Placement Status'] == 'Hired']
hired_students
```

```python
grouped_hired = hired_students.groupby(['Batch Name', 'Company Name']).size().reset_index(name='Count')
grouped_hired
```

```python
companies_grouped = grouped_hired.groupby('Company Name').sum()
companies_grouped
```

# COMPANY WISE HIRED PIE CHART

```python
company = ['GroupM', 'IIDE', 'Kinnect', 'Schbang']
students_hired = [15, 18, 28, 14]

plt.figure(figsize = (11,6))
plt.pie(students_hired, labels = company, autopct = '%0.1f%%', startangle = 140, explode = [0,0,0,0], colors = ['r', 'g', 'b', 'y'])
plt.title("Companies Wise Hired Students")
plt.legend(loc = 4)
plt.show()
```

#Observations:
* As per the observation we have seen that kinnect has the company who's hired the most of the students.
* 28 students hired by the kinnect company
* Percenatge of hiring students is 37.3% by the kinnect

# To analyze out of the total hired students - what percentage were Interns & Full Timer (Add record count and percentage).

#Find out under which salary package were the maximum number of students hired

```python
salary_package_counts = hired_students['Yearly Salary Cluster'].value_counts()
salary_package_counts
```

```python
max_salary_package = salary_package_counts.idxmax()
max_students_hired = salary_package_counts[max_salary_package]
```

```python
print("Maximum number of students were hired under the salary package:", max_salary_package)
print("Number of students hired under this package:", max_students_hired)
```

#Observation:
* Maximum number of students were hired under the salary package: 4 Lakh - 5 Lakh
Number of students hired under this package: 18
